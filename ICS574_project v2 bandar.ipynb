```python
# verify there are no missing values in the specific sectors dataset
print("Specific Sectors Dataset Missing Values:")
print(stocks_specific_df.isnull().sum())
```

```python
# print the rows with missing values
print("Specific Sectors Dataset Missing Values Rows:")
print(stocks_specific_df[stocks_specific_df.isnull().any(axis=1)])
```

```python
# print the total number of rows in the specific sectors dataset
print("Specific Sectors Dataset Total Rows:")
print(stocks_specific_df.shape)
```

```python
# divide the missing values by the total number of rows to get the percentage of missing values
# we see it is a very small percentage and we can ignore it during the analysis
# there are various reasons for missing values in the dataset, that could range from stock trading being halted for a day, or the stock not being listed on that day by Tadawul, etc.
print("Specific Sectors Dataset Missing Values Percentage:")
print(stocks_specific_df.isnull().sum() / stocks_specific_df.shape[0] * 100)
```

```python
# plot the closing price of the stocks and brent dataset
plt.figure(figsize=(10, 6))
plt.title("Stocks Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price (SAR)")
plt.plot(stocks_brent_df['Date'], stocks_brent_df['close'])
plt.plot(stocks_brent_df['Date'], stocks_brent_df['Price'])
plt.legend(['Stocks', 'Brent'])
plt.show()
```

```python
# plot the closing price of the stocks and brent dataset with different colors for each sector and brent
plt.figure(figsize=(10, 6))
plt.title("Stocks Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price (SAR)")
sns.lineplot(x='Date', y='close', data=stocks_brent_df, hue='sectoer')
sns.lineplot(x='Date', y='Price', data=stocks_brent_df, color='black')
plt.show()
```

```python
# plot the correlation between the stocks and brent dataset
plt.figure(figsize=(10, 6))
plt.title("Stocks Closing Prices")
sns.heatmap(stocks_brent_df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
# create pairplot for the stocks and brent dataset
plt.figure(figsize=(10, 6))
plt.title("Stocks Closing Prices")
sns.pairplot(stocks_brent_df,hue='sectoer')
plt.show()
```
